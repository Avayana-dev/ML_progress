```python
# Day 5: Week 1 Completion
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

print("WEEK 1 COMPLETE - Kaggle ML Certificate Earned")

# Quick skills demo
np.random.seed(42)
data = pd.DataFrame({
    'feature_1': np.random.randint(1, 100, 50),
    'feature_2': np.random.randint(50, 150, 50),
    'target': np.random.randint(10000, 50000, 50)
})

X = data[['feature_1', 'feature_2']]
y = data['target']

model = RandomForestRegressor(n_estimators=10, random_state=42)
model.fit(X, y)
predictions = model.predict(X)
mae = mean_absolute_error(y, predictions)

print(f"Final Model MAE: ${mae:,.2f}")
print("Week 1: 5 notebooks, ML fundamentals, certificate earned")
print("Next: Python certificate, then real projects post-exams")
```
