```python
#day 2 : Model Validation Practice 
#what I learned Today :MAE ,Train-test split , model validation

import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split

print("Day 2 Practice: model validation concepts")
```

```python
#creating the realistic house data for practice

np.random.seed(42)
size = np.random.randint(800,2500,50)
bedrooms = np.random.randint(1,5,50)
age = np.random.randint(1,30,50)

#price formula : base +size*100 =bedrooms * 10000 - age*500 + noise

price = 50000 +size*100 + bedrooms*10000 - age*500 +np.random.randint (-10000,10000,50)

housing_data = pd.DataFrame ({
    'house_size' : size,
    'bedrooms' : bedrooms,
    'house_age' : age,
    'price' : price
})

print ("Sample Housing Data Created: ")
print(housing_data.head())
print(f"\nDataset shape: {housing_data.shape}")
```

```python
#Lesson 3 Concept : Spilliting data for validation
x = housing_data [['house_size','bedrooms','house_age']]
y = housing_data ['price']
x_train, x_test,y_train, y_test = train_test_split(x,y,test_size = 0.2, random_state=42
                                          )
print("Train-Test split results")
print(f"training set : {x_train.shape[0]} houses")
print (f"Testing set : {x_test.shape[0]} houses  ")

print(f"features used : {list(x.columns)} ")
```

```python
# Training the model on training data
model = DecisionTreeRegressor(random_state=42)
model.fit(x_train,y_train)

# Making predictions on test data
test_predictions = model.predict(x_test)

# Calculating error by using MAE
mae = mean_absolute_error(y_test, test_predictions)

print("These are the Model Validation Results:")
print(f"Mean Absolute Error: ${mae:,.2f}")
print(f"Average house price: ${y.mean():,.2f}")
print(f"Error as % of average price: {(mae/y.mean()*100):.1f}%")
```

```python
# Simple Prediction Comparison
print("Prediction Comparison:")
print("Actual Price  __ Predicted Price")
print("-" * 35)

for i in range(5):  # Show only first 5
    actual = y_test.iloc[i]
    predicted = test_predictions[i]
    print(f"${actual:,.0f}    |  ${predicted:,.0f}")

print(f"\n Average Error: ${mae:,.2f}")
```
